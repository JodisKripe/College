```python
import random
ans = random.randint(1, 10)

guessed = False
no_of_guesses = 0

print("Welcome to the guessing game! You have 3 attempts to guess the number.")

while not guessed and no_of_guesses < 3:
    guess = int(input((f"Guess {no_of_guesses+1}: ")))
    no_of_guesses += 1
    if guess == ans:
        guessed = True
    else:
        print("Sorry, your guess is incorrect.")

if guessed:
    print("Congratulations! You guessed the number.")
else:
    print("Out of guesses, you lose.")
```
